```python
import pandas as pd
from openpyxl import Workbook
import requests
import json
import time

df = pd.read_excel('fornecedores.xlsx')

cnpjs = df.iloc[:,2]
wb = Workbook()
ws = wb.active
ws.title = "CNAE"

wb1 = Workbook()
ws1 = wb1.active
ws1.title = "Emails"
n1 = 1

for cnpj in cnpjs:
    time.sleep(2)
    cnpj = str(cnpj)
    cnpj = cnpj.replace(".","")
    cnpj = cnpj.replace("-","")
    cnpj = cnpj.replace("/","")
    print("Voltei o for")

    try:
        resultado = requests.get(f"https://api.cnpjs.dev/v1/{cnpj}")
        if resultado.status_code == 200:
            resultado = resultado.json()
            cnaePrincipal = resultado['cnae_fiscal_principal']
            cnaeSecundario = resultado['cnae_fiscal_secundaria']
            #print(f"Código: {cnaePrincipal['codigo']}\nNome: {cnaePrincipal['nome']}")
            qntCnpj = 0
            for i in cnaeSecundario:
                qntCnpj += 1

            qntCnpj += 2
            
            n2 = 1
            while(True):
                if(n2 == 1):
                    ws.cell(row=n1, column=n2, value=cnpj)
                    ws1.cell(row=n1, column=n2, value=cnpj)
                    n2 += 1

                    ws.cell(row=n1, column=n2, value=cnaePrincipal['codigo'])
                    ws1.cell(row=n1, column=n2, value=resultado['email'])
                    n2+=1

                    ws.cell(row=n1, column=n2, value=cnaePrincipal['nome'])
                    n2+=1

                    ws.cell(row=n1, column=n2, value=0)
                    n1+=1

                    print("Fiz o primeiro")

                elif(n2 == 4):
                    for i in cnaeSecundario:
                        n2 = 1
                        nome = str(i)
                        nome = nome.replace("{'codigo': ","")
                        nome = nome.replace("'}","")
                        nome = nome.split(", 'nome': '")
                        print(nome)
                        ws.cell(row=n1, column=n2, value=cnpj)
                        ws1.cell(row=n1, column=n2, value=cnpj)
                        n2+=1
                        ws.cell(row=n1, column=n2, value=i['codigo'])
                        n2+=1
                        ws.cell(row=n1, column=n2, value=i['nome'])
                        n2+=1
                        ws.cell(row=n1, column=n2, value=1)
                        n2+=1
                        n1+=1
                else:
                    break
        wb.save('cnae.xlsx')       
        wb1.save('email.xlsx')       
        print(f'{qntCnpj} - {n1} - {cnpj}')       
    except:
        print("Error")
        pass
```
